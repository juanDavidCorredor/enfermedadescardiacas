```python
import matplotlib.pyplot as plt
import pandas as pd

sex_mapping = {0: 'Femenino', 1: 'Masculino'}
sex_distribution = airbnb_data['sex'].map(sex_mapping).value_counts()

sex_distribution.plot(kind='bar', color=['skyblue', 'lightcoral'])
plt.title('Sexo')
plt.xlabel('Sexo')
plt.ylabel('Cantidad')
plt.grid(axis='y', alpha=0.75)
```
